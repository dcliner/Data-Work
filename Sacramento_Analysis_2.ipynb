```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
Sacramento_df = pd.read_csv("sacramento.csv")
```

```python
Sacramento_df.head()
```

```python
Sacramento_df.columns
```

```python
Sacramento_df.drop('Unnamed: 0', axis= 1, inplace= True)
```

```python
Sacramento_df.head()
```

```python
dummy = pd.get_dummies(Sacramento_df['baths'])
```

```python
dummy.head(25)
```

```python
dummy[1.0].head()
```

```python
Bath = dummy[1.0]
```

```python
Sacramento_df_2 = pd.concat((Sacramento_df,Bath), axis=1)
```

```python
Sacramento_df_2.head()
```

```python
Sacramento_df_2.drop('baths',axis= 1, inplace= True)
```

```python
Sacramento_df_2.head()
```

```python
Sacramento_df_2.rename(columns={1.0: 'baths'}, inplace= True)
```

```python
Sacramento_df_2.head(23)
```

```python
Xtrain = Sacramento_df_2[['beds','sqft','price']].values
Ytrain = Sacramento_df_2[['baths']].values
```

```python
X = sm.add_constant(Xtrain)
```

```python
mod = sm.Logit(Ytrain, X).fit()
```

```python
print(mod.params.round(2))
print(mod.pvalues.round(2))
print('The smallest p-value is for sqft')
```
